#### Building the recommendation engine
Having all the data in shape, it's finally time to build the recommendation engine itself. This step will assume for the moment a fixed and arbitrary number of features. This is just to be sure that I have implemented everything in the proper way. Later on, I will implement hyperparameter tuning to find the optimal number of features.

First step is to build the model itself by a dedicated function

```python
# define global parameters of use
NUMBER_OF_USERS    = music_habits_dataset['encoded_user_name'].nunique()
NUMBER_OF_ARTISTS  = music_habits_dataset['encoded_artist_msid'].nunique()
NUMBER_OF_FEATURES = int(numpy.round(NUMBER_OF_ARTISTS**(0.25))) # naive empirical estimate 

def build_recommendation_engine():    
    # Create input layer for user names
    user_name_input = keras.layers.Input(shape=(1,))
    # Embedd users with the number of features 
    embedded_user_name = keras.layers.Embedding(NUMBER_OF_USERS, 
                                                NUMBER_OF_FEATURES,
                                                embeddings_initializer='he_normal',
                                                name = 'user_preferences')(user_name_input)
    # reshape embedding layer to correspond to qualitative picture of matrix factorization
    embedded_user_name = keras.layers.Reshape((NUMBER_OF_FEATURES,))(embedded_user_name)
    
    # Create input layer for artist msid
    artist_msid_input = keras.layers.Input(shape=(1,))
    # Embedd users with the number of features 
    embedded_artists_msid = keras.layers.Embedding(NUMBER_OF_ARTISTS, 
                                                   NUMBER_OF_FEATURES,
                                                   embeddings_initializer='he_normal',
                                                   name = 'artist_features')(artist_msid_input)
    # reshape embedding layer to correspond to qualitative picture of matrix factorization
    embedded_artists_msid = keras.layers.Reshape((NUMBER_OF_FEATURES,))(embedded_artists_msid)

    # compute an approximation of the original values of log_number_of_features
    # using that the two embeddings are the matrix factorization of log_number_of_features
    appriximated_log_number_of_features = keras.layers.Dot(axes=1)([embedded_user_name, embedded_artists_msid])

    # build recommendation engine a keras model
    recommendation_engine = keras.Model(inputs  = [user_name_input, artist_msid_input], 
                                        outputs = appriximated_log_number_of_features)
    recommendation_engine.compile(loss = 'mean_squared_error', optimizer = 'adam')

    return recommendation_engine

# run function to build engine
recommendation_engine = build_recommendation_engine()
# visual summary of the model
recommendation_engine.summary()
```

With the recommendation engine model at hand, we can now train it and evaluate its performances.

```python
recommendation_engine.fit(x = x_train, 
                          y = y_train, 
                          batch_size = 64, 
                          epochs = 10,
                          verbose = 1, 
                          validation_data = (x_validate, y_validate))
```

```python
# define functions to use for the recommendation
def find_artists_not_in_library(music_dataset, user_name):
    """
    This function finds the artists_msid values that are not in the library of 
    user with name user_name using the collections of user libraries in music_dataset
    Inputs: 
      music_dataset: 
        A pandas dataframe that contains how many times a user has listened to a
        given artist over a given span of time. 
      user_name:
        String with the name of the user.
    Outputs:
        A numpy array with the msid of artists that do not appear in the user library 
    """
    # isolate entries from single user
    log_of_listens = music_dataset[music_dataset['user_name'] == user_name]
    
    # print human readable artists in the user library 
    print('Artists in ' + user_name + ' library:')
    print(log_of_listens['artist_name'].unique())
    
    # find unique artist_msid in user library
    unique_msid_in_library = log_of_listens['artist_msid'].unique()
    
    # find unique artist_msid not in user library
    unique_msid_not_in_library = set(music_dataset['artist_msid'].unique()) - set(unique_msid_in_library)
    unique_msid_not_in_library = list(unique_msid_not_in_library)
    return (unique_msid_not_in_library)




def encode_user_name_and_artist_not_in_library(user_name, 
                                               unique_msid_not_in_library,
                                               user_encoder,
                                               artist_encoder):
    """
    This function takes the user specified in user_name and the list of artists
    not in the user library, and it formats them into arrays that can be
    passed to the tensorflow recommendation engine.
    Inputs:
      user_name:
        String with the name of the user.
      unique_msid_not_in_library:
        List of artists_msid values that are not in the user library as ouputed
        by find_artists_not_in_library.
      user_encoder, artist_encoder:
        sklear.().LabelEncoder() objects to encode user_name and unique_msid_not_in_library
        into a format that can be passed "as is" to the tensorwlof.model object recommendation_engine
    Outputs:
      user_name_tensorflow_input, artist_not_in_library_tensorflow_input:
        numpy arrays with the encoded values of user_name and unique_msid_not_in_library
        using the encoders user_encoder, artist_encoder respectively.
    """
    # encode user name directly in format suitabele for tensorflow model
    encoded_user_name = user_encoder.transform([user_name])
    user_name_tensorflow_input = numpy.array([encoded_user_name[0] for i in unique_msid_not_in_library])
    # encode artist msid not in user library in format suitabele for tensorflow model
    artist_not_in_library_tensorflow_input = artist_encoder.transform(unique_msid_not_in_library)
    return user_name_tensorflow_input, artist_not_in_library_tensorflow_input




def recommend_artists(music_dataset,
                      user_name, 
                      user_encoder,
                      artist_encoder,
                      recommendation_engine,
                      number_of_suggestions):
    """
    this function recommends a number number_of_suggestions artists to user with name 
    user_name usign recommendation_engine.
    Inputs:
      music_dataset: 
        A pandas dataframe that contains how many times a user has listened to a
        given artist over a given span of time.
      user_name:
        String with the name of the user.
      user_encoder, artist_encoder:
        sklear.().LabelEncoder() objects to encode user_name and unique_msid_not_in_library
        into a format that can be passed "as is" to the tensorwlof.model object recommendation_engine
      recommendation_engine:
        Tensorflow model object that has been trained as a recommendation model using
        matrix factorization based on implicit features.
      number_of_suggestions:
        Integer that specifies the number of suggestions to give among all the possible ones
    Outputs:
      recommended_artists:
        List of the top number_of_suggestions recommended artists for user_name
        in a human readable format.
    """
    # find artists not in user library
    unique_msid_not_in_user_library = find_artists_not_in_library(music_habits_dataset, a_user)                   

    # encode user name and artists not in user libray in tensorflow ready format
    user_name_tf, artist_not_in_library_tf = encode_user_name_and_artist_not_in_library(a_user, 
                                                                                        unique_msid_not_in_user_library,
                                                                                        user_encoder,
                                                                                        artist_encoder)
    # predict scores for artists not in user library
    scores = recommendation_engine.predict([user_name_tf, artist_not_in_library_tf])
    scores = [item for sublist in scores for item in sublist]
    
    # find best number_of_suggestions recommendations
    recommendations = numpy.argsort(scores)[len(scores) - number_of_suggestions:]
    recommendation_msids = artist_encoder.inverse_transform(recommendations)
    
    all_artists = music_dataset[['artist_name', 'artist_msid']].drop_duplicates() 
    recommended_artists = all_artists[all_artists['artist_msid'].isin(recommendation_msids)]
    
    # print human readable recommended artists
    print('\nRecommended for ' + user_name + ':')
    print(recommended_artists['artist_name'])
    
    return recommended_artists
```
